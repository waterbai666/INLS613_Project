```python
def pickCate(catelist, data):
    cate = []
    product = []
    review_star = []
    review = []
    cost = []
    tag = []
    
    for i in range(len(data)):
        if data.iloc[i].category in catelist and len(data.iloc[i].REVIEWLIST)>0:
            reviews = data.iloc[i].REVIEWLIST.split("}, {")
            reviews[0]=reviews[0][2:]
            reviews[-1]=reviews[-1][:-2]
            
            for j in range(len(reviews)):
                if len(reviews[j])>0:
                    dic=ast.literal_eval("{"+reviews[j]+"}")
                    review.append(dic['review_body'])
                    if len(dic['review_star'])>0:
                        star = dic['review_star'][26]
                    else:
                        star = ""
                    review_star.append(star)
                    if star == "4" or star == "5":
                        tag.append("positive")
                    elif star == "1" or star == "2":
                        tag.append("negative")
                    else:
                        tag.append("fair")
             
                    cate.append(data.iloc[i].category)
                    product.append(data.iloc[i].product_name)
                    cost.append(data.iloc[i].cost)
   
    redic = {'cate':cate,'product':product,'cost':cost,'review_star':review_star,'review':review,'tag':tag}
    df = pd.DataFrame(redic)
    return df       

    
```

```python
for i in train.category.unique():
    pickCate([i],train).to_csv("data/cate/"+i+".csv")
```
